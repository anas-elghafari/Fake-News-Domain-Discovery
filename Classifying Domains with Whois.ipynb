```python
def store_whois_into_dicts(sources, label, whois_dict, whois_dict_iana):
    failed = []
    for s in sources:
        print("Source now is: ", s)
        if (s, label) in whois_dict:
            continue
        try:
            w =  pythonwhois.get_whois(s.strip())
        except Exception as e:
            print("Exception when obtaining whois with no specified server")
            print(e, "\n\n")
            failed.append(s)
            continue
        try:
            w_iana = pythonwhois.get_whois(s.strip(), server='whois.iana.org')
        except Exception as e:
            print("Exception when obtaining whois from whois.iana.org")
            print(e, "\n\n")
            failed.append(s)
            continue
        whois_dict[(s, label)] = w
        whois_dict_iana[(s, label)] = w_iana
        #time.sleep(1)
    print("Failed to get WHOIS on ", len(failed), " domains")
    print("\n".join(failed))
    return failed
        
        
store_whois_into_dicts(fn_sources, 'fake', whois_dict, whois_dict_iana)
store_whois_into_dicts(msm_sources, 'msm', whois_dict, whois_dict_iana)
```

```python
def get_kv_strings(kprefix, d):
    result = []
    if hasattr(d, "keys"):
        for k, v in d.items():
            if k =="raw":
                continue
            else:
                result.extend(get_kv_strings(kprefix+ "/" +k, v))
    elif isinstance(d, list):
        for e in d:
            result.append(kprefix + " : " + str(e))
    else:
        result.append(kprefix + " : " + str(d))
    return result


def get_kv_from_raw(whois_raw):
    result = []
    w = whois_raw.split("\n")
    w = [l.strip() for l in w]
    for l in w:
        if l.startswith("URL of"):
            print("Reached disclaimer: ", l)
            break
        else:
            result.append(l)
    return result
```

```python
featvecs = []
featvecs_json = []

for k, v in whois_dict.items():
    v_iana = whois_dict_iana[k]
    print("Domain: ", k)
    vec = [k]
    vec.extend(get_kv_from_raw(v['raw'][0]))
    featvecs.append(vec)
    vec2 = []
    vec2.extend(get_kv_strings("", v))
    vec2.extend(get_kv_strings("", v_iana))
    vec2 = sorted(set(vec2))
    vec2.insert(0, k)
    featvecs_json.append(vec2)
```

```python
def extract_prop_names(fv):
    prop_names = []
    for l in fv[1:]:
        #print(l)
        #print(len(l))
        l = l.split(":")
        prop_names.append(l[0].strip())
    return prop_names


propnames = []
for v in featvecs_json:
    propnames.extend(extract_prop_names(v))
```

```python
days_nums_strings = {0:"Monday", 1:"Tuesday", 2:"Wednesday", 3:"Thursday",4:"Friday", 5:"Saturday", 6:"Sunday"}

def enrich_vector(fv):
    newv = []
    dateval = ""
    for val in fv[1:]:
        val = val.lower()
        if val.startswith("/creation_date"):
            dateval = parse(val[val.index(": ")+2:])
            year = int(dateval.year)
            newv.append("/creation_year: " + str(year))
            newv.append("/domain older than 2 years: " + str(year<2016))
            newv.append("/domain older than 5 years: " + str(year<2013))
            newv.append("/domain older than 10 years: " + str(year<2008))
            weekday = dateval.weekday()
            newv.append("/creation weekday: " + days_nums_strings[weekday])
            newv.append("/created on weekend: " + str(weekday in [5,6]))
            hour = dateval.hour
            newv.append("/created outside typical business hours: " + str((hour>18 or hour <8)))
            print("creation date: ", dateval)
            
        if val.startswith("/updated_date"):
            dateval = parse(val[val.index(": ")+2:])
            weekday = dateval.weekday()
            newv.append("/update weekday: " + days_nums_strings[weekday])
            newv.append("/updated on weekend: " + str(weekday in [5,6]))
            hour = dateval.hour
            newv.append("/updated outside typical business hours: " + str((hour>18 or hour <8)))
            print("update date: ", dateval)
            
        if val.startswith("/nameservers"):
            nameserver = val[val.index(": ")+2:]
            nameserver_domain = nameserver.split(".")
            nameserver_domain = ".".join(nameserver_domain[-2:])
            newv.append("/nameserver_domain: "+ nameserver_domain.upper())
            
        if val.find("@")>-1:
            #look at the email domain
            email_domain = val[val.index("@")+1:]
            newv.append("/email_domain: "+ email_domain)
            
    newv = sorted(set(newv))  #get rid of duplicates, especially from nameserver normalization
    print("newv:\n", "\n".join(newv))
    ret = [x for x in fv]
    ret.extend(newv)
    return ret
            
    
    
enriched_vecs = []    
for fv in featvecs_json:
    print("\ncurrent domain: ", fv[0])
    enriched_vecs.append(enrich_vector(fv))
```

```python
freq = defaultdict(lambda: [0, 0])
for v in enriched_vecs:
    for val in v[1:]:
        if v[0][1] == "msm":
            freq[val][0] += 1
        elif v[0][1] == "fake":
            freq[val][1] += 1
        else:
            print("Unknown value: ", v[1])
```

```python
#sorting features by mutual information. Most informative features (in either direction) first:
freqsorted = sorted([i for i in freq.items()], key= lambda x: abs(x[1][3]- x[1][2]), reverse=True)
print("\n\n".join([str(x) for x in freqsorted[0:200]]))
```

```python
def select_dimensions(instances):
    features = defaultdict(lambda: 0)
    for current_instance in instances:
        for d in set(current_instance[1:]):
            features[d] +=1
            
    print("\n\nNumber of instances: ", len(instances))
    print("Total number of possible features: ", len(features))
    selected_features = sorted(features.keys())
    return selected_features


def insts_to_binary_vecs(insts, dimensions):
    vecs = np.zeros(shape = [len(insts), len(dimensions)], dtype='uint')
    for i, inst in enumerate(insts):
        print("building vector for instance:", inst[0])
        vec = np.array([0 if not d in inst[1:] else 1 for d in dimensions])
        vecs[i] = (vec)
    return vecs 

d = select_dimensions(enriched_vecs)
print("\n".join(d))
bvecs = insts_to_binary_vecs(enriched_vecs, d)

#adding back the label and id, after the vector was tuned binary:
instance_vectors = []
names = [x[0] for x in enriched_vecs]
print(names)
for name, bvec in zip(names, bvecs):
    inst = [name]
    inst.extend(bvec)
    instance_vectors.append(inst)
```

```python
def crosseval(insts):
    confusion_mat = defaultdict(lambda: [])
    random.shuffle(insts)
    testsize = math.floor(0.1*len(insts))
    print("test size for each fold: ", testsize)
    for foldnum in range(10):
        print("Now evaluating on fold: ", foldnum+1)
        testfold = insts[foldnum*testsize: (foldnum+1)*testsize]
        trainfold = [x for x in insts if not x in testfold]
        y_train = [inst[0][1] for inst in trainfold]
        x_train = [inst[1:] for inst in trainfold]
        y_test = [inst[0][1] for inst in testfold]
        x_test = [inst[1:] for inst in testfold]
        ids = [inst[0][0] for inst in testfold]
        print("ids of test fold: ", ids)
        confusion_mat = eval_1fold(x_train, y_train, x_test, y_test, ids, confusion_mat)
        print("\n\n")
        for k, v in confusion_mat.items():
            print(k, len(v))
            
    return confusion_mat
    


def eval_1fold(x_train, y_train, x_test, y_test, ids, confusion_mat):
    m = linear_model.LogisticRegression(C=1e5)
    t1 = datetime.datetime.now()
    m.fit(x_train, y_train)
    t2 = datetime.datetime.now()
    print("Finished training on ", len(x_train), " instances in: ", t2-t1)
    y_pred = m.predict(x_test)
    for i, (realval, predval) in enumerate(zip(y_test,y_pred)):
        recons_vec = [(ids[i], realval)]
        recons_vec.extend(x_test[i])
        confusion_mat[(realval, predval)].append(recons_vec)
        
    return confusion_mat


confusion = crosseval(instance_vectors)
```

```python
def prec_recall(conf):
    class_size = len(conf[('fake', 'fake')]) + len(conf[('fake', 'msm')])
    labeled_fake = len(conf[('fake', 'fake')]) + len(conf[('msm', 'fake')])
    true_positives = len(conf[('fake', 'fake')])
    false_positives = len(conf[('msm', 'fake')])
    false_negatives = len([x for x in conf[('fake', 'msm')]])
    #print("true pos:", true_positives, false_positives, false_negatives, labeled_fake, class_size)
    print("precision: ", float(true_positives/labeled_fake))
    print("recall: ", true_positives/float(class_size))
    
prec_recall(confusion)
```

```python
def get_trained_model(instances):
    y_train = [inst[0][1] for inst in instances]
    x_train = [inst[1:] for inst in instances]
    m = linear_model.LogisticRegression(C=1e5)
    t1 = datetime.datetime.now()
    m.fit(x_train, y_train)
    t2 = datetime.datetime.now()
    print("Training the model finished in:", t2-t1)
    return m

def get_prediction(model, domain_name):
    w =  pythonwhois.get_whois(domain_name)
    w_iana = pythonwhois.get_whois(domain_name.strip(), server='whois.iana.org')
    vec = ["/domain: " + domain_name]
    vec.extend(get_kv_strings("", w))
    vec.extend(get_kv_strings("", w_iana))
    vec = sorted(set(vec))
    enriched_v = enrich_vector(vec)
    bvec = insts_to_binary_vecs([enriched_v], d)
    pred = model.predict(bvec)
    return pred
    
    
domainname = "breitbart.com"
model = get_trained_model(instance_vectors)
pred = get_prediction(model, domainname)
print("Prediction for domain \"" + domainname + "\" is:", pred[0])
```
